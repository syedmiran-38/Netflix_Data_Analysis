```python
data.head(2)
```

```python
data['Date_N'].dt.year.value_counts()
```

```python
data['Date_N'].dt.year.value_counts().plot(kind= 'bar', color = 'red')
plt.title('Number of TV shows and Movies per Year', weight='bold').set_fontsize('15')
plt.gcf().set_size_inches(12, 8)
plt.show()
```
